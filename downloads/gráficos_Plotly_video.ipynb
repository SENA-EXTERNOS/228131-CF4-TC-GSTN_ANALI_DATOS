**Gráficos dinámicos con plotly**

**Histograma**

```python
fig = px.histogram(df, x='edad', y='sexo', histfunc='count',title='Cantidad de hombres vrs Cantidad de mujeres')
fig.show()
```

```python
import plotly
plotly.offline.plot(fig,filename='histograma_hombres_vrs_mujeres.html')
```

**Rango =valor maximo-valor minimo**

```python
rango=df['edad'].max()-df['edad'].min()
rango
```

**Cuartiles**

```python
minimo=df['edad'].min()
Q1=df['edad'].quantile(q=0.25)
Q2=df['edad'].quantile(q=0.5)
Q3=df['edad'].quantile(q=0.75)
maximo=df['edad'].quantile(q=1)
print(minimo,Q1,Q2,Q3,maximo)
```

```python
df['edad'].describe()
```

```python
iqr=Q3-Q1
iqr
```

**Deteccion de datos atipicos(outliers)**<br>
**intervalo**
**(Q1-1.5*iqr,Q3+1.5*iqr)**

```python
minlimite=Q1-1.5*iqr
```

```python
maxlimite=Q3+1.5*iqr
```

```python
print(minlimite,maxlimite)
```

**Gráfico de cajas y bigotes con plotly**

**Se observa Q1:cuartil 1,Q2:cuartil 2(mediana),Q3:cuartil 3, valor mínimo , valor máximo**

```python
fig = px.box(df, x='edad',title='Edad')
fig.show()
```

```python
fig = px.box(df, x='estrato',y='edad',title='Estrato vrs Edad')
fig.show()
```

```python
agrupamiento=df.groupby(['estrato','edad']).agg({'monto_prestado':'sum',})
agrupamiento
```

```python
agrupamiento=agrupamiento.reset_index()
agrupamiento
```

```python
import altair as alt
```

```python
alt.Chart(agrupamiento).mark_bar().encode(
    x='estrato',
    y='monto_prestado',
    color='edad',opacity=alt.value(0.8)
).properties(width=220)
```
